```python
import math

# Task 3: Probability - Immigration Analysis
print("\n" + "="*50)
print("TASK 3: PROBABILITY ANALYSIS")
print("="*50)

# Poisson distribution parameter
immigration_rate = 5

print(f"Immigration Model: Poisson Distribution (λ = {immigration_rate})")
print("-" * 50)

# Calculate distribution properties
mean_events = immigration_rate
variance_events = immigration_rate
std_deviation = math.sqrt(immigration_rate)

print("Statistical Properties:")
print(f"Expected events per year: {mean_events}")
print(f"Variance: {variance_events}")
print(f"Standard deviation: {std_deviation:.4f}")

# Calculate key probabilities
prob_5_events = (math.e**(-immigration_rate) * immigration_rate**5) / math.factorial(5)

print(f"\nProbability Analysis:")
print(f"P(exactly 5 events) = {prob_5_events:.4f}")
print(f"Expected range: {mean_events - std_deviation:.1f} to {mean_events + std_deviation:.1f} events")

# Show distribution for first few values
print(f"\nProbability Distribution:")
for k in range(0, 8):
    prob_k = (math.e**(-immigration_rate) * immigration_rate**k) / math.factorial(k)
    print(f"P(X = {k}) = {prob_k:.4f}")
```
