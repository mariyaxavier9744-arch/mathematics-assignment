```python
# Task 2: Calculus - Growth Rate Analysis
import numpy as np

print("\n" + "="*50)
print("TASK 2: CALCULUS ANALYSIS")
print("="*50)

# Population parameters
initial_pop = 1000
growth_coefficient = 0.03
time_period = 20

# Define time points for analysis
time_points = [0, 5, 10, 15, 20]

print("Population Function: P(t) = 1000 × e^(0.03t)")
print("Growth Rate Function: P'(t) = 30 × e^(0.03t)")
print("\nTime Series Analysis:")
print("-" * 40)
print("Year\tPopulation\tGrowth Rate")
print("-" * 40)

max_growth = 0
final_population = 0

for t in time_points:
    population_t = initial_pop * np.exp(growth_coefficient * t)
    growth_rate_t = growth_coefficient * population_t

    if t == time_period:
        final_population = population_t
        max_growth = growth_rate_t

    print(f"{t}\t{population_t:.0f}\t\t{growth_rate_t:.2f}")

print(f"\nMaximum Growth Analysis:")
print(f"Final population (t=20): {final_population:.1f}")
print(f"Maximum growth rate: {max_growth:.2f} people/year")
print(f"Total growth: {((final_population/initial_pop - 1) * 100):.0f}%")
```
