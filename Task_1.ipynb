```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import math

# Task 1: Linear Algebra - Eigenvalue Analysis
print("="*50)
print("TASK 1: LINEAR ALGEBRA ANALYSIS")
print("="*50)

# Define transition matrix
transition_matrix = np.array([[1.2, 0.1],
                              [0.05, 1.1]])

print("Population Transition Matrix:")
print(f"A = {transition_matrix}")

# Calculate eigenvalues and eigenvectors using numpy
eigenvalues, eigenvector_matrix = np.linalg.eig(transition_matrix)

# Sort eigenvalues in descending order
sorted_indices = np.argsort(eigenvalues.real)[::-1]
lambda1, lambda2 = eigenvalues[sorted_indices]
eigvec1, eigvec2 = eigenvector_matrix[:, sorted_indices].T

# Normalize eigenvectors (second component = 1)
eigvec1_normalized = eigvec1 / eigvec1[1]
eigvec2_normalized = eigvec2 / eigvec2[1]

print(f"\nResults:")
print(f"λ₁ = {lambda1.real:.6f}")
print(f"λ₂ = {lambda2.real:.6f}")
print(f"Eigenvector 1: ({eigvec1_normalized[0].real:.3f}, 1)")
print(f"Eigenvector 2: ({eigvec2_normalized[0].real:.3f}, 1)")

# Stability analysis
print(f"\nStability Analysis:")
if lambda1.real > 1 and lambda2.real > 1:
    print("Both eigenvalues > 1: System shows exponential growth")
    print(f"Dominant growth factor: {max(lambda1.real, lambda2.real):.3f}")
```
