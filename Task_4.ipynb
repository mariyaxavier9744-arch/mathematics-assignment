```python
import numpy as np
import scipy.stats as stats
import math

# Task 4: Statistics - District Analysis
print("\n" + "="*50)
print("TASK 4: STATISTICAL ANALYSIS")
print("="*50)

# District population dataset
district_populations = [1050, 1070, 1065, 1080, 1045, 1075, 1060, 1070]
sample_size = len(district_populations)

print("District Population Data:")
for i, pop in enumerate(district_populations, 1):
    print(f"District {i}: {pop} residents")

# Calculate sample statistics
sample_mean = np.mean(district_populations)
sample_variance = np.var(district_populations, ddof=1)
sample_std = np.sqrt(sample_variance)

print(f"\nSample Statistics:")
print(f"Sample size: {sample_size}")
print(f"Mean population: {sample_mean:.2f}")
print(f"Standard deviation: {sample_std:.4f}")
print(f"Variance: {sample_variance:.4f}")

# Confidence interval calculation
confidence_percent = 95
alpha_level = 0.05
degrees_of_freedom = sample_size - 1

# Get t-critical value
t_critical = stats.t.ppf(1 - alpha_level/2, degrees_of_freedom)
standard_error = sample_std / math.sqrt(sample_size)
margin_of_error = t_critical * standard_error

# Calculate interval bounds
lower_limit = sample_mean - margin_of_error
upper_limit = sample_mean + margin_of_error

print(f"\n{confidence_percent}% Confidence Interval:")
print(f"Degrees of freedom: {degrees_of_freedom}")
print(f"t-critical: {t_critical:.4f}")
print(f"Standard error: {standard_error:.4f}")
print(f"Margin of error: {margin_of_error:.2f}")
print(f"Confidence interval: ({lower_limit:.2f}, {upper_limit:.2f})")

print(f"\nInterpretation:")
print(f"We are {confidence_percent}% confident that the true mean district")
print(f"population lies between {lower_limit:.0f} and {upper_limit:.0f} residents")
```
